```python
def get_week_of_month(query_date):
    """
    獲取某個日期是當月的第幾周
    """
    end = int(query_date.strftime("%W"))
    begin = int(datetime.datetime(query_date.year, query_date.month, 1).strftime("%W"))
    return end - begin + 1
```
