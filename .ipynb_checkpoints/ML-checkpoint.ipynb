```python
#CH3_8.py
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
x = np.arange(start = -4,stop = 5)
y = np.arange(start = -4,stop = 5)
X,Y = np.meshgrid(x,y)
#建立子圖
Z = X + Y
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'},figsize = (20,10))
ax.plot_wireframe(X,Y,Z,color = 'g')
ax.scatter(X,Y,Z,color = 'b')
ax.set_title('會製3D網格圖')
ax.set_xlabel('X軸',color = 'b')
ax.set_ylabel('Y軸',color = 'b')
ax.set_zlabel('Z軸',color = 'b')
```

```python
#CH3_9.py
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
x = np.linspace(-5,5,30)
y = np.linspace(-5,5,30)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
#建立一個新的圖像
fig = plt.figure(figsize = (20,10))
#建立第1個子圖(3D網格圖)
ax1 = fig.add_subplot(2,2,1,projection = '3d')
ax1.plot_wireframe(X,Y,Z,linewidth = 0.5,cmap = 'rainbow')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('3D網格圖')
#建立第二個子圖(3D表面圖)
ax1 = fig.add_subplot(2,2,2,projection = '3d')
ax1.plot_surface(X,Y,Z, cmap = 'rainbow')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('3D表面圖')
#建立第三個子圖(3D等高線)
ax2 = fig.add_subplot(2,2,3)
contour = ax2.contour(X,Y,Z,cmap = 'rainbow')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('等高線圖')
fig.colorbar(contour)
#建立第四個子圖(填充等高線圖)
ax2 = fig.add_subplot(2,2,4)
contour = ax2.contourf(X,Y,Z,cmap = 'rainbow')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('填充等高線圖')
fig.colorbar(contour)
#顯示圖像
plt.subplots_adjust(wspace = 0.4,hspace = 0.4)#子圖間距設定
plt.show()
```

```python
#Ch4_2.py
import matplotlib.pyplot as plt
from sympy import Symbol,solve
import numpy as np
plt.rcParams['font.family'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
a = Symbol('a')
b = Symbol('b')
eq1 = 600*a + b
eq2 = 1000*a + b -12
ans = solve((eq1,eq2))
print(f'a = {ans[a]}')
print(f'b = {ans[b]}')
pt_x1 = 600
pt_y1 = ans[a]*pt_x1 + ans[b]
pt_x2 = 1000
pt_y2 = ans[a]*pt_x2 + ans[b]
x = np.linspace(0,2500,500)
y = ans[a]*x + ans[b]
plt.plot(x,y)
plt.plot(pt_x1,pt_y1,'-o')
plt.text(pt_x1+60 ,pt_y1-2,'P1')
plt.plot(pt_x2,pt_y2,'-o')
plt.text(pt_x2 +60,pt_y2-2,'P2')

plt.grid()
```

```python
#CH4_5.py
import matplotlib.pyplot as plt
import numpy as np
from sympy import Symbol,solve

plt.rcParams['font.family'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False

x = Symbol('x')
y = Symbol('y')

eq1 = x + y -35
eq2 = 2*x + 4*y -100
ans = solve((eq1,eq2))
print(f'雞 = {ans[x]}')
print(f'兔 = {ans[y]}')
line1_x = np.linspace(0,100,100)
line1_y = [35-z for z in line1_x]
line2_x = np.linspace(0,100,100)
line2_y = [25 - 0.5*q for q in line2_x]
plt.plot(line1_x,line1_y)
plt.plot(line2_x,line2_y)
plt.plot(ans[x],ans[y],'-o')
plt.text(ans[x]-5,ans[y]+5,'('+str(ans[x])+','+str(ans[y])+')')
plt.xlabel('雞')
plt.ylabel('兔')
plt.grid()
```

```python
#CH5_1.py
import math
film = [5,7,8,10,2] #玩命關頭特徵值
film_titles = ['復仇者聯盟','決戰中途島','冰雪奇緣','雙子殺手']
film_features = [[2,8,8,5,6],[5,6,9,2,5],[8,2,0,0,10],[5,8,8,8,3]]
dist = []
for f in film_features:
    distances = 0
    for i in range(len(f)):
        distances +=(film[i] - f[i])**2
    dist.append(math.sqrt(distances))
min_ = min(dist)
min_index = dist.index(min_)
print(f'與玩命關頭最相似電影：{film_titles[min_index]}')
print(f'相似度值：{dist[min_index]:6.2f}')
for i in range(len(dist)):
    print(f'影片:{film_titles[i]},相似度：{dist[i]:6.2f}')
```
